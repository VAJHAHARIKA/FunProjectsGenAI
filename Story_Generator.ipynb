Story Generator: This tiny project is a local text generator using a small language model (GPT-2). It takes a prompt from you (like "Once upon a time") and then auto-generates the rest of the story or sentence — all without needing the internet or an API key.



```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")
prompt = input("Enter a prompt: ")

result = generator(prompt, max_length=50, num_return_sequences=1)
print(result[0]['generated_text'])
```
